```python
figsize(7, 5)
bb = np.linspace(0, 260, 100)
colors = ['b', 'r', 'g', 'k', 'y']
print len(bb)
k=0
nh_b = np.zeros(100)
for j in range(1, 10, 2):
    for i in range(100):
        nh_b[i] = nh(260.0, bb[i], j)
    plt.plot(bb, nh_b, lw=2, c=colors[k], label='c='+str(j))
    k+=1
    yscale('log')
plt.ylabel(r'$\mathrm{N_H(cm^{-2})}$', fontsize=25)
legend(ncol=3, fontsize=15, loc='lower center')
plt.xlabel(r'$\mathrm{Impact\ parameter\ b(kpc)}$', fontsize=25)
savefig('NH-b.png', bbox_inches='tight')
```
